# Jitterplot Gallery

```python
import numpy as np
import matplotlib.pyplot as plt
import scprep
```

```python
values = np.concatenate([np.random.uniform(0, 1, 300),
                         np.random.normal(0.5, 0.5, 300),
                         np.random.beta(0.5, 0.5, 300)])
labels = np.concatenate([np.repeat('Uniform(0, 1)', 300),
                         np.repeat('Normal(0.5, 0.5)', 300),
                         np.repeat('Beta(0.5, 0.5)', 300)])
```

```python
scprep.plot.jitter(labels, values, c=values, title="Continuous color vector")
```

```python
scprep.plot.jitter(labels, values, c=labels, title="Discrete color vector")
```

```python
scprep.plot.jitter(labels, values, c=np.round(values, 1), discrete=True, legend=False, 
                   title="Explicitly discrete color vector")
```

```python
scprep.plot.jitter(labels, values, c=np.round(values), discrete=False, legend=False, 
                   title="Explicitly continuous color vector")
```

```python
scprep.plot.jitter(labels, values, c=values, cmap=['red', 'blue'],
                      title="Continuous color vector, list colormap")
```

```python
scprep.plot.jitter(labels, values, c=values, vmin=0.25, vmax=0.75,
                      title="Continuous color vector, specified range")
```

```python
scprep.plot.jitter(labels, values, c=values, vmin=0, cmap_scale='sqrt',
                      title="Continuous color vector, nonlinear scale")
```

```python
scprep.plot.jitter(labels, values, c=labels, 
                   cmap={'Uniform(0, 1)' : 'darkred',
                         'Normal(0.5, 0.5)' : 'darkblue',
                         'Beta(0.5, 0.5)' : 'darkgreen'},
                   title="Discrete color vector, dictionary colormap")
```

```python
scprep.plot.jitter(labels, values, c=values, 
                   cmap=scprep.plot.colors.tab10_continuous(n_colors=3),
                   title="Continuous color vector, graded discrete colormap")
```

```python
scprep.plot.jitter(labels, values, c=values, 
                   cmap=scprep.plot.colors.tab40(),
                   title="40-element discrete colormap")
```

```python
scprep.plot.jitter(labels, values, c='xkcd:light periwinkle', title="Named color")
```

```python
scprep.plot.jitter(labels, values, c=np.random.uniform(0, 1, [len(values), 4]), title="RGBA color vector")
```

```python
scprep.plot.jitter(labels, values, c=values, title="Custom axis ticks",
                   xticks=False, yticks=[-1, 1, 2], yticklabels=['a', 'b', 'c'])
```

```python
scprep.plot.jitter(labels, values, c=values, plot_means=False, title="Don't plot mean points")
```

```python
scprep.plot.jitter(labels, values, c=values, means_s=200, means_c="green", title="Custom mean point size/color")
```

```python
scprep.plot.jitter(labels, values, c=values, sigma=0.05, title="Custom jitter width")
```

```python
scprep.plot.jitter(labels, values, c=values, legend_title="Colorbar/legend title")
```

```python
scprep.plot.jitter(labels, values, c=values, fontsize=8, title="Custom font size")
```

```python
scprep.plot.jitter(labels, values, c=values, figsize=(12, 4), title="Custom figure size")
```

```python
scprep.plot.jitter(labels, values, c=values, filename="file.png", dpi=300, title="Save output to file")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 4))
scprep.plot.jitter(labels, values, c=labels, cmap=['darkred', 'darkblue', 'darkgreen'], ax=ax1)
scprep.plot.jitter(labels, values, c=values, cmap=['red', 'blue'], ax=ax2, colorbar=False)
scprep.plot.tools.generate_colorbar(cmap=['red', 'blue'], ax=[ax2, ax1], 
                                    vmin=np.min(values), vmax=np.max(values))
plt.suptitle("Colorbar with multiple equally spaced axes")
```
